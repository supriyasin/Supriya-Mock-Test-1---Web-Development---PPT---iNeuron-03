```python
#3. Explain what is execution context in detail with diagram

"""In programming, an execution context refers to the environment in which a piece of code is executed. It includes all 
   the variables, functions, and other resources that are accessible during the execution of that code. Each time a 
   function is called or a block of code is executed, a new execution context is created.

   To understand execution contexts better, let's break down the different components and their relationships using a diagram.
   
   Execution Context
-----------------------------------------------------
|                    Variable Environment            |
|---------------------------------------------------|
| Variables                                          |
| Functions                                          |
| This                                               |
-----------------------------------------------------
|                    Scope Chain                     |
|---------------------------------------------------|
| Outer Environment                                  |
| Outer Environment                                  |
| Outer Environment                                  |
-----------------------------------------------------
|                    This Value                      |
-----------------------------------------------------


   1. Variable Environment: The variable environment contains all the declared variables and functions within the current 
      execution context. It's essentially a container for all the identifiers (variable and function names) and their 
      associated values. The variable environment also includes the special variable "this," which refers to the object
      that is currently being executed.

   2. Scope Chain: The scope chain represents the hierarchical chain of outer environments or scopes that are accessible 
      from the current execution context. It allows variables and functions to be resolved from outer scopes. The scope 
      chain is used to look up variables that are not found in the current execution context.

   3. Outer Environment: Each execution context has a reference to its outer environment or scope. The outer environment 
      represents the variables and functions of the parent or enclosing scope. This allows nested functions to access 
      variables defined in their containing functions or global scope.

   4. This Value: The "this" value refers to the object that the current code is being executed upon. It provides a 
      context for the execution and is dynamically determined based on how a function is invoked.
      
   To illustrate the concept, let's consider an example:
   
   function outer() {
  var x = 10;
  
  function inner() {
    var y = 20;
    console.log(x + y);
  }
  
  inner();
}

outer();


   In this example, when the outer() function is called, a new execution context is created for it. The variable 
   environment of this context will include the variable x with a value of 10. The scope chain will contain a
   reference to the outer environment, which is the global scope.

   Next, when the inner() function is called within the outer() function, a new execution context is created for it. 
   The variable environment of this context will include the variable y with a value of 20. The scope chain will 
   contain references to both the outer environment of inner() (which is outer()'s variable environment) and the 
   global scope.

   By understanding the execution context and its components, developers can determine how variables are accessed,
   how function scopes are resolved, and how the "this" keyword behaves within their code. It provides a crucial 
   understanding of the runtime environment and helps in managing and debugging code effectively."""
```
